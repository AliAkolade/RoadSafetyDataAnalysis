```python
def age_band(age_of_casualty): 
    if age_of_casualty <= 5:
        return '0-5' 
    elif age_of_casualty <= 10:
        return '6-10' 
    elif age_of_casualty <= 15:
        return '11-15' 
    elif age_of_casualty <= 20:
        return '16-20'
    elif age_of_casualty <= 25:
        return '21-25'
    elif age_of_casualty <= 35:
        return '26-35'
    elif age_of_casualty <= 45:
        return '36-45'
    elif age_of_casualty <= 55:
        return '46-55'
    elif age_of_casualty <= 65:
        return '56-65'
    elif age_of_casualty <= 75:
        return '66-75'
    else:
        return 'Over 75'
```

```python
df['age_group'] = df['age_of_casualty'].apply(age_band)
```

```python
def cas_severity(casualty_severity):
    if casualty_severity == 1:
        return 'Fatal'
    elif casualty_severity == 2:
        return 'Serious'
    else:
        return "Slight"
```

```python
df['cas_sev'] = df['casualty_severity'].apply(cas_severity)
```

```python
df.head()[['casualty_severity','cas_sev']]
```

```python
df=df.drop('casualty_severity', axis=1)
```

```python
df.shape
```

```python
X = df.drop('cas_sev', axis=1)
X = pd.get_dummies(X,drop_first=True)
```

```python
#X
```

```python
y = df['cas_sev']
```

```python
#y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

```python
preds = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test,preds))
```

```python
from sklearn.metrics import confusion_matrix

# ConfusionMatrixDisplay(dt,X_test,y_test)
cm = confusion_matrix(y_test, preds, labels=dt.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dt.classes_)
disp.plot()
plt.show()
```

```python
imp_feats = pd.DataFrame(data=dt.feature_importances_,index=X.columns,columns=['Feature Importance']).sort_values("Feature Importance")
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=imp_feats.sort_values('Feature Importance'),x=imp_feats.sort_values('Feature Importance').index,y='Feature Importance')
plt.xticks(rotation=90)
plt.title("Feature Importance for Decision Tree");
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(14,10),dpi=150)
plot_tree(dt,filled=True,feature_names=X.columns);
```
